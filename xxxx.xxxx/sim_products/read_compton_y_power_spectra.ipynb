```python
#import modules
import numpy as np, os
from pylab import *
```

```python
rcParams['figure.dpi'] = 150
```

```python
#read file
fname = 'compton_y_power_spectra.npy'
pspec_dict = np.load(fname, allow_pickle = True).item()
for expkey in pspec_dict:
    print(expkey)
    el, cl_mean, cl_median = pspec_dict[expkey] #mean/median from 500 simulations
```

```python
#plot
color_dict = {'Fiducial': 'black', 'CMB-HD': 'darkred', 'S4-Ultra Deep': 'orangered', \
'S4-Wide': 'darkgreen', 'SPT-3G': 'goldenrod', 'SO-Goal': 'lightskyblue', 'SO-Baseline': 'royalblue', \
'SPTpol': 'navy', 'SPT-SZ': 'slateblue'}

ax = subplot(111, yscale = 'log')
for expkey in pspec_dict:
    el, cl_mean, cl_median = pspec_dict[expkey] #mean/median from 500 simulations
    dl_fac = el * (el+1)/2/np.pi
    #plot median spectrum
    plot(el, dl_fac * cl_median, color = color_dict[expkey], label = expkey)
xlabel(r'Multipole $\ell$')
ylabel(r'$\ell (\ell+1)C_{\ell}^{yy}/2\pi$')
xlim(200., 10000.)
ylim(2e-14, 2e-12)
legend(loc = 4)
show()
```
