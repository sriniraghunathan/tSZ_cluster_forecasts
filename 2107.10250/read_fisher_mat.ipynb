```python
#modules
import numpy as np, os, glob, re
```

```python
#get folder name and files
opfolder = 'data_products/fisher/'
fisher_flist = sorted( glob.glob('%s/*.npy' %(opfolder)) )
print(fisher_flist)
```

```python
desired_params_to_print = ['vir_model_1_etav', 'vir_model_2_Av', 'vir_model_2_Bv', 'one_minus_hse_bias', 'm_nu', 'w0']
for fname in fisher_flist:
    vir_eff_model = re.findall(r'vir_model[0-9]*', fname)[0].replace('vir_model','')
    opline = '\nVirialisation model = %s\n' %(vir_eff_model)
    print(opline)
    fisher_dict = np.load(fname, allow_pickle=True).item()
    for which_fisher in fisher_dict: #CMB or cluster or combined
        opline = '\tType = %s' %(which_fisher)
        print(opline)
        for expname in fisher_dict[which_fisher]:
            opline = '\t\tExperiment = %s' %(expname)
            print(opline)
            param_names, F_mat, C_mat = fisher_dict[which_fisher][expname]
            errors = np.diag(C_mat)**0.5
            for p, e in zip(param_names, errors):
                if p not in desired_params_to_print: continue
                print('\t\t\t%s: %g' %(p, e))
```
